 Polymorphism by uing sample dataset which is related to salary

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dfsal=pd.read_csv(r"E:\Feature engineering\Data Set set\linear Regression\Linear Regression\Polynomial\salary_poly.csv")
dfsal.drop('Position',axis=1,inplace=True)
dfsal
```

```python
x=dfsal[['Level']]
y=dfsal[['Salary']]
display(x,y)
```

```python
plt.scatter(x,y,color = "yellow")
plt.show
```

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg
```

```python
display(lreg.fit(x,y))
```

```python
print(lreg.score(x,y))
```

```python
lreg.predict(x)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lreg.predict(x),color="green")
plt.show()
```

```python
# polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg= PolynomialFeatures()
poly_reg

poly_reg.fit(x)
xpoly= poly_reg.transform(x)
print(xpoly)
```

```python
lreg1= LinearRegression()
lreg1
print(lreg1.fit(xpoly,y))
lreg1.score(xpoly,y)
```

```python
lreg1.predict(xpoly)
```

```python
# compare  1st predict bias line and new polynomial predict bias line
plt.scatter(x,y,color="yellow")
plt.plot(x,lreg.predict(x),color="green")
plt.plot(x,lreg1.predict(xpoly),color="lightgreen")
plt.show()
```

```python
# polynomial  regression with degree 
poly_reg_degree= PolynomialFeatures(degree=3)
poly_reg_degree.fit(x)
xpoly2=poly_reg_degree.transform(x)
lreg2=LinearRegression()
lreg2.fit(xpoly2,y)
lreg2.score(xpoly2,y)
```

```python
lreg2.predict(xpoly2)
```

```python
plt.scatter(x,y,color='yellow')
plt.plot(x,lreg.predict(x),color="green")
plt.plot(x,lreg1.predict(xpoly),color="lightgreen")
plt.plot(x,lreg2.predict(xpoly2))
plt.show()
```

```python
plt.scatter(x,y,color='yellow')
plt.plot(x,lreg.predict(x),color="green")
plt.show()
```

```python
plt.scatter(x,y,color='yellow')
plt.plot(x,lreg1.predict(xpoly),color="lightgreen")
plt.show()
```

```python
plt.scatter(x,y,color='yellow')
plt.plot(x,lreg2.predict(xpoly2))
plt.show()
```

```python
# checking increasing the degree

poly_reg_degree= PolynomialFeatures(degree=4)
poly_reg_degree.fit(x)
xpoly3=poly_reg_degree.transform(x)
lreg3=LinearRegression()
lreg3.fit(xpoly3,y)
lreg3.score(xpoly3,y)
```

```python
lreg3.predict(xpoly3)
```
